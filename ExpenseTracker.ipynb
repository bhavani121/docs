```python
from datetime import datetime
import csv
import os


expenseList=[]
expenseDictionary={}
monthlyBudget=0


def initiate_tracker():
    print(f'{name}, What would you like to do today \n 1.Add Expense \n 2.View expenses \n 3.Track Budget \n 4.Save expenses \n 5.Exit the program \n Enter the number to choose')
    inputValidated=True;
    while inputValidated ==True:
        choiceNumber=input()
        try:
            choiceNumber = float(choiceNumber)
            inputValidated=False;
        except:
                print('enter a number')
    return choiceNumber
        
def show_menu():
    print("\n--- Personal Expense Tracker ---")
    print("1. Add expense")
    print("2. View expenses")
    print("3. Track budget")
    print("4. Save expenses")
    print("5. Exit")


                
def validate(dateEntered):
     try:
        datetime.strptime(dateEntered, '%Y-%m-%d')
        return True
     except ValueError:
        return False

def currentMonthEntry(row):
    if(datetime.now().month == datetime.strptime(row['expenseDate'], "%Y-%m-%d").month):
        print('returning True')
        return True
    else:
        return False

    
    
def addEntryToCSV(dataDictionary):

      file_exists = os.path.isfile('expense.csv')
    
      with open('expense.csv', mode='a', newline='') as file:
        fieldnames = ['expenseDate', 'category', 'amount', 'description']
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        
        # Write header only if file doesn't exist
        if not file_exists:
            writer.writeheader()
        
        writer.writerow(dataDictionary)

    
def add_expense():
    print(f'{name}, you have choosen option Add Expense')
    expenseDate=input('enter the day of expense in YYYY-MM-DD format')
    while(validate(expenseDate) == False):
            expenseDate= input('enter the day of expense in YYYY-MM-DD format')
    category=input('enter the category of the expense')
    try:
          amount=float(input('enter the amount of expense done'))
    except:
            print('enter the expense amount in numbers')
    description=input('give a brief description of the expense made')
    expenseDictionary={'expenseDate':expenseDate,"category":category,"amount":amount,"description":description}
    expenseList.append(expenseDictionary)
    return expenseDictionary
        
        
def view_expenses():
    print(f'{name}, you have choosen option View Expense')
    if not expenseList:
        print("there are no expenses to show")
    else:
        for i ,exp in enumerate(expenseList,1):
            if 'expenseDate' in exp and 'category' in exp and 'amount' in exp and 'description' in exp:
                print(f"{i}. Date:{exp['expenseDate']},Category:{exp['category']},Amount:Rs{exp['amount']:.2f},Description:{exp['description']}")
           
            else:
                print('Incomplete entry,not listing')
    

def  load_expenses():
    if not os.path.exists('expense.csv'):
        return

    with open('expense.csv', mode='r') as file:
        reader = csv.DictReader(file)
        print('rread file')
        for row in reader:
            try:
                expense = {
                    'expenseDate': row['expenseDate'],
                    'category': row['category'],
                    'amount': float(row['amount']),
                    'description': row['description']
                }
                expenseList.append(expense)
                print(expense)
            except (KeyError, ValueError):
                print("Skipped an invalid row in CSV.")

def track_budget():
        print(f'{name}, you have choosen option Track Budget')
        totalAmount=0;
        budgetForMonth=input(f'{name}, please enter your budget for current  month')
        fileRead=open('expense.csv','r')
        reader=csv.DictReader(fileRead)
        for row in reader:
            if(currentMonthEntry(row)):
                amount = float(row['amount'])
                totalAmount=amount+totalAmount
        if totalAmount>=float(budgetForMonth):
            print('you have reached your budget limit for this month')
            print('you have crossed the budget by', totalAmount-float(budgetForMonth))
        if(totalAmount<float(budgetForMonth)):
            print('you have',float(budgetForMonth)-totalAmount,'Rs left for this month')

def save_expenses():
   addEntryToCSV(add_expense())
    


def main():
    global name
    name=input('Enter your name')
    load_expenses()
    while True:
        show_menu()
        choice= input('choose numbers between 1-5')
        match float(choice):
            case 1:
                add_expense()
            case 2:
                view_expenses()
            case 3:
                track_budget()
            case 4:
                save_expenses()
            case 5:
                view_expenses()
                print("Goodbye! Have a great day.")
                break
            case default:
                return 'No option choosen.Exiting the program'
       
        
   


if __name__ == "__main__":
    main()

        
        
        
        
        
    
            
    
    
       
```
